```python
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
import seaborn as sns
```

```python
path= "H:\\study\\Association Rules\\Assignments\\book.csv"
```

```python
df = pd.read_csv(path)
```

```python
df_ap = apriori(df,min_support=0.072,max_len=4,use_colnames = True)
```

```python
df_ap
```

```python
df_ar = association_rules(df_ap,metric='lift',min_threshold=1)
```

```python
df_ar[df_ar['confidence']> 0.7]
```

Cross-Selling can be improved by combining products - items
- The book shop layout can be changed so that sales can be improved when certain items are kept together.
- Promotional activities which are an advertising campaign can be carried out to increase the sales of books that customers do not buy.
- Collective discounts can be offered on these books if the customer buys both of them.
